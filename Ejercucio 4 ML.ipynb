```
import matplotlib.pyplot as plt
import numpy as np
```

```
n=7#Numero de n Vectores(hasta 7).
Vectores= [np.array([np.cos(t*2*np.pi/8), np.sin(t*2*np.pi/8)]) for t in range(n+2)]#Definimos el conjunto de venctores
#y definimos de donde a donde va n, en este caso n=[1,10]
M=[[2.5, -2], [-2,3]]#Definimos la matriz.
sigma = np.array(M)#Lo volvemos un objeto numpy
S = np.array(Vectores)#Lo volvemos un objeto numpy
sigma_inver = np.linalg.inv(sigma)#Invertimos sigma(la matriz)
sigma_x = np.array([np.dot(sigma, S[i, :]) for i in range(len(Vectores))])
sigmainver_x = np.array([np.dot(sigma_inver, S[i, :]) for i in range(len(Vectores))])
```

```
#Graficamos.
plt.plot(S[:, 0], S[:, 1], 'o', color = 'b')
plt.plot(sigma_x[:, 0], sigma_x[:, 1], '-o', color = 'g')
plt.plot(sigmainver_x[:, 0], sigmainver_x[:, 1], '-o', color = 'r')
plt.ylabel('Eje de las y')
plt.xlabel('Eje de las x')
plt.title("Vectores "+ r"$\mathbf{x}$")
plt.grid()
plt.show()
```
